# Approximation of functions

```python
import numpy as np
import matplotlib.pyplot as pyplot
import scipy.linalg
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
A = np.array([[1, 1],[1, 15]])
B = np.array([f(1),f(15)])
w =scipy.linalg.solve(A, B)
```

```python
x = np.arange(0,15,0.1).reshape(1,-1)
y = w[0]+w[1]*x
```

```python
pyplot.plot(x,y, 'go--')
pyplot.show()
```

```python
A = np.array([[1, 1, 1],[1, 8,64],[1,15,225]])
B = np.array([f(1),f(8),f(15)])
w = scipy.linalg.solve(A, B)
```

```python
x = np.arange(0,15,0.1).reshape(1,-1)
y = w[0]+w[1]*x+w[2]*x**2
```

```python
pyplot.plot(x,y, 'go--')
pyplot.show()
```

```python
A = np.array([[1, 1, 1,1],[1, 4,16,64],[1,10,100,1000],[1,15,225,3375]])
B = np.array([f(1),f(4),f(10),f(15)])
w = scipy.linalg.solve(A, B)
```

```python
x = np.arange(0,15,0.1).reshape(1,-1)
y = w[0]+w[1]*x+w[2]*x**2+w[3]*x**3
```

```python
pyplot.plot(x,y, 'go--')
pyplot.show()
```
